```python
# ================================
# STEP 1: Import Libraries
# ================================
print("STEP 1: Importing Libraries")

import pandas as pd
import matplotlib.pyplot as plt

# ================================
# STEP 2: Load Dataset
# ================================
print("\nLoading Dataset")

df = pd.read_csv("Iris.csv")

print("\nFirst 5 Rows:")
print(df.head())

# ================================
# STEP 3: Basic Info
# ================================
print("\nShape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

# ================================
# STEP 4: Data Cleaning Checks
# ================================
print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

# ================================
# STEP 5: Statistical Summary
# ================================
print("\nStatistical Summary:")
print(df.describe())

# ================================
# STEP 6: Target Distribution
# ================================
print("\nTarget Variable Distribution:")
print(df["Species"].value_counts())

# ================================
# STEP 7: Histograms
# ================================
print("\nHistogram Grid")

df.hist(figsize=(10, 8), bins=15)
plt.suptitle("Histogram of Features")
plt.show()

# ================================
# STEP 8: Skewness
# ================================
print("\nSkewness:")
print(df.skew(numeric_only=True))

# ================================
# STEP 9: Boxplots (Bivariate Analysis)
# ================================
print("\nBoxplots by Species")

features = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]

for feature in features:
    df.boxplot(column=feature, by="Species", figsize=(6,4))
    plt.title(f"{feature} vs Species")
    plt.suptitle("")
    plt.show()

# ================================
# STEP 10: Correlation Matrix
# ================================
print("\nCorrelation Matrix")

corr_matrix = df.drop("Species", axis=1).corr()
print(corr_matrix)

# ================================
# STEP 11: Heatmap
# ================================
print("\nCorrelation Heatmap")

plt.figure(figsize=(6,5))
plt.imshow(corr_matrix, cmap='coolwarm')
plt.colorbar()

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

plt.title("Correlation Heatmap")
plt.show()

# ================================
# STEP 12: Outlier Detection
# ================================
print("\nOutlier Detection")

df.drop("Species", axis=1).boxplot(figsize=(8,5))
plt.title("Outlier Detection")
plt.show()
```
